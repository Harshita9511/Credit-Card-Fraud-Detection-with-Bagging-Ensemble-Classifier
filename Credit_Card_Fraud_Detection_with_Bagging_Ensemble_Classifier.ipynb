## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('creditcard.csv')
```

```python
df.head()
```

```python
df.describe()
```

## Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
df['Scaled_Amount']=scaler.fit_transform(df['Amount'].values.reshape(-1,1))
df['Scaled_Time']=scaler.fit_transform(df['Time'].values.reshape(-1,1))
```

```python
df.head()
```

```python
df=df.drop(['Amount', 'Time'], axis=1)
df.head()
```

## Dealing with Unbalanced Data

```python
sns.countplot(df['Class'], palette='viridis')
df['Class'].value_counts()
```

```python
df=df.sample(frac=1)
```

```python
fraud= df.loc[df['Class'] == 1]
non_fraud=df.loc[df['Class']==0][:492]
```

```python
df = pd.concat([fraud, non_fraud])
```

```python
data = df.sample(frac=1, random_state=42)
print(data.shape)
data
```

## Exploratory Data Analysis

```python
sns.countplot(data['Class'], palette='spring')
data['Class'].value_counts()
```

```python
plt.figure(figsize=(20,14))
sns.heatmap(data.corr(), annot=True, cmap='viridis')
```

## Train Test Split

```python
X=data.drop('Class', axis=1)
y=data['Class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

## Train the Bagging Classifier

```python
from sklearn.ensemble import BaggingClassifier
```

```python
clf = BaggingClassifier(n_estimators=10, random_state=0)
```

```python
clf.fit(X_train, y_train)
```

## Predictions and Evaluations

```python
predictions= clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test, predictions))
print('\n')

labels=[0,1]
cmx=confusion_matrix(y_test, predictions, labels)
fig = plt.figure()
ax = fig.add_subplot()
cax = ax.matshow(cmx)
plt.title('Confusion matrix of the classifier')
fig.colorbar(cax)
ax.set_xticklabels([''] + labels)
ax.set_yticklabels([''] + labels)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

print('\n')
print(classification_report(y_test, predictions))
```
